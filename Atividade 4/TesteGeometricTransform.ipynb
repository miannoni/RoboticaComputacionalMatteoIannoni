## Exemplo de uso da função getAffineTransform da OpenCV



Atenção:  o que este exemplo tem a ver com a atividade 1?

A função `getAffineTransform` recebe 3 correspondências de pontos entrada-saída e produz uma matriz de transformação. Isso é análogo ao que vocês precisam fazer na atividade. Em outras palavras, vocês não vão usar esta função, mas farão um trabalho parecido.

**Depois**  que vocês obtiverem a matriz de transformação, para conferir se está certa, vocês podem usar a função `cv2.warpAffine()` passando a matriz que obtiveram e conferir se o resultado é parecido com a matriz de saída que forneci.




```python

import cv2
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

img = cv2.imread('messi5.jpg')
rows,cols = img.shape[:2]

# Source points
srcTri = np.array([(0,0),(cols-1,0),(0,rows-1)], np.float32)

# Corresponding Destination Points. Remember, both sets are of float32 type
dstTri = np.array([(cols*0.0,rows*0.33),(cols*0.85,rows*0.25), (cols*0.15,rows*0.7)],np.float32)

# Affine Transformation
warp_mat = cv2.getAffineTransform(srcTri,dstTri)   # Generating affine transform matrix of size 2x3
dst = cv2.warpAffine(img,warp_mat,(cols,rows))     # Now transform the image, notice dst_size=(cols,rows), not (rows,cols)

# Image Rotation
center = (cols/2,rows/2)                           # Center point about which image is transformed
angle = 50.0                                      # Angle, remember negative angle denotes clockwise rotation
scale = 0.6                                        # Isotropic scale factor.

rot_mat = cv2.getRotationMatrix2D(center,angle,scale) # Rotation matrix generated
dst_rot = cv2.warpAffine(img,rot_mat,(cols,rows))     # Now transform the image wrt rotation matrix

rot_mat
```
